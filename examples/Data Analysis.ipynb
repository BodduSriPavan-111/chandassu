```python
import os

import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.getcwd()
```

```python
path= 'c:\\Users\\bsnra\\Desktop\\chandassu\\dataset'
os.chdir( path )
```

```python
df= pd.DataFrame()
for i in os.listdir():
    if df.shape[0]==0:
        df= pd.read_excel(i)
    else:
        df= pd.concat([df, pd.read_excel(i)])

    if "geethamu" in df["type"].value_counts().index:
        print(i)
```

```python
df.describe()
```

##### satakam

```python
satakam= df["satakam"].value_counts()
satakam
```

```python
plt.figure( figsize= (15, 5) )

plt.barh( range(satakam.shape[0]), satakam.values, color= "cornflowerblue" )
plt.axvline(x=100, color='lime', linestyle='--', label='Non Vupajaathi Satakam')
plt.axvline(x=200, color='green', linestyle='--', label='Vupajaathi Satakam')

plt.yticks( range(satakam.shape[0]), satakam.index )
plt.title("Distribution of no.of Padyams from each Satakam")
plt.legend(
            title= "Satakam Type",
            loc='upper right',
            bbox_to_anchor=(1.2, 1)
           )

plt.show()
```
