```python
df= pd.DataFrame()
for i in os.listdir():
    if df.shape[0]==0:
        df= pd.read_excel(i)
    else:
        df= pd.concat([df, pd.read_excel(i)])
```

##### satakam

```python
satakam= df["satakam"].value_counts()
satakam
```

```python
plt.figure( figsize= (15, 5) )

plt.barh( range(satakam.shape[0]), satakam.values, color= "silver" )
plt.axvline(x=100, color='lime', linestyle='--', label='Vertical Line')

plt.yticks( range(satakam.shape[0]), satakam.index )
plt.title("Distribution of no.of Padyams from each Satakam")

plt.show()
```

##### bhedam

```python
bhedam= df["bhedam"].value_counts()
bhedam
```

```python
plt.figure( figsize= (15, 5) )
plt.bar( range(bhedam.shape[0]), bhedam.values )
plt.xticks( range(bhedam.shape[0]), bhedam.index  )
plt.show()
```

```python
plt.figure( figsize= (15, 5) )
plt.pie( bhedam.values, labels= bhedam.index )
plt.show()
```
